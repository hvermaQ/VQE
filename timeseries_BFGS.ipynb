```python
import pickle
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
#with open('global_noisy_seeds_gates_init/noiseless_COBY_seeds_init.pkl', 'rb') as file:
with open('global_noisy_seeds_BFGS_tightcov_20seeds/noiseless_BFGS_seeds_init.pkl', 'rb') as file:
    yruns = pickle.load(file)
```

```python
dep = np.arange(1, 11, 1, dtype = int)
nseeds = 20
```

```python
res_data_idm = []
res_data_ids = []
val_list_idm = []
val_list_ids = []
for pt in dep:
    yran = []
    ylis = []
    for ht in np.arange(0, nseeds, 1, dtype = int):
        #re = eval(yruns['%s'%ht][pt-1].meta_data['optimization_trace'])
        re = eval(yruns['%s'%ht][pt-1].meta_data['int_energy'])
        vv = yruns['%s'%ht][pt-1].value
        app = [vv for yt in range(1000-len(re))]
        yran.append(re + app)
        ylis.append(vv)
    yseed_opt = np.array(yran)
    yseed_val = np.array(ylis)
    mean_y = np.mean(yseed_opt, axis=0)
    std_y = np.std(yseed_opt, axis=0)
    val = np.mean(yseed_val, axis=0)
    val_sig = np.std(yseed_val, axis=0)
    res_data_idm.append(mean_y)
    res_data_ids.append(std_y)
    val_list_idm.append(val)
    val_list_ids.append(val_sig)
```

```python
#selected depth
sel = 1
```

```python
font = {'family' : 'Tahoma',
        'size'   : 12}

matplotlib.rc('font', **font)
```

```python
# Plotting
x = np.arange(1, 1001, 1, dtype = int)
plt.figure(figsize=(10, 6))

# Plot each individual run
for i in range(nseeds):
    re = eval(yruns['%s'%i][sel].meta_data['int_energy'])
    vv = yruns['%s'%i][sel].value
    app = [vv for yt in range(1000-len(re))]
    yran = re + app
    plt.plot(x, yran, color='grey', alpha=0.5)

# Plot the mean line
plt.plot(x, res_data_idm[sel], color='blue', label='Mean')
# Compute upper and lower bounds for shading
upper_bound = res_data_idm[sel] + res_data_ids[sel]
lower_bound = res_data_idm[sel] - res_data_ids[sel]
# Add shading for the range (mean ± std)
plt.fill_between(x, lower_bound, upper_bound, color='green', alpha=0.2, label='Mean ± 1 Std Dev')

plt.xlim([1, 50])
# Add labels and legend
plt.xlabel('t')
plt.ylabel('E(t)')
#plt.title('Shaded Time Series Plot')
plt.legend()
plt.savefig('layer_5_timeseries.pdf', bbox_inches = 'tight')
# Show plot
plt.show()
```
